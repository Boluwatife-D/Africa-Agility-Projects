### START OF DATA CLEANING PROCESS

```python
#To give a count of the total number of missing values for each column in the dataset sorted in lowest to highest order.
exit_data.isnull().sum().sort_values()
```

```python
#Selecting specific columns from the dataset and passing them as a dataframe into a new variable while making a copy so it does now alter the original data
new_ed1 = exit_data[['ID', 'SeparationType', 'Position', 'Region', 'Business Unit', 'Employment Status', 'Job dissatisfaction', 'Gender', 'Age']].copy()
```

```python
new_ed1
```

```python
#To check the information of the new dataframe
new_ed1.info()
```

```python
#To check which of the seleceted columns have missing values.
new_ed1.isnull().sum().sort_values()
```

```python
#Since the business unit column has a lot of missing values we would be dropping the business unit colum from our dataset.
new_ed2 = new_ed1.drop(['Business Unit'], axis = 1)
new_ed2
```

```python
#To fill the missing entries in each column with the most occuring entry for that column (i.e. mode since the data is classified categorically)
new_ed2['Position'] = new_ed2['Position'].fillna(value = new_ed2['Position'].mode()[0])
new_ed2['Employment Status'] = new_ed2['Employment Status'].fillna(value = new_ed2['Employment Status'].mode()[0])
new_ed2['Age'] = new_ed2['Age'].fillna(value = new_ed2['Age'].mode()[0])
new_ed2['Gender'] = new_ed2['Gender'].fillna(value = new_ed2['Gender'].mode()[0])
```

```python
new_ed2
```

```python
#To confirm if our data is now free of missing values.
new_ed2.isnull().sum().sort_values()
```
